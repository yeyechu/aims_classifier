## 문서 분류기 EDA


### 1. 패키지 및 데이터 불러오기

```python
import matplotlib.pyplot as plt

from set_font_matplot import set_nanumgothic_font
set_nanumgothic_font()
from analyzer import analyze_dataset_train, analyze_dataset_test
```

```python
train_dir = '../../data/train'
test_dir = '../../data/test'
```

```python
train_class_counts, train_widths, train_heights, train_total_images = analyze_dataset_train(train_dir)
test_widths, test_heights, test_total_images = analyze_dataset_test(test_dir)
```

```python
print("Train Dataset:")
print(f"Total Images: {train_total_images}")
print(f"Class Counts: {dict(train_class_counts)}")
print(f"Average Width: {sum(train_widths) / len(train_widths):.2f}")
print(f"Average Height: {sum(train_heights) / len(train_heights):.2f}")

print("\nTest Dataset:")
print(f"Total Images: {test_total_images}")
print(f"Average Width: {sum(test_widths) / len(test_widths):.2f}")
print(f"Average Height: {sum(test_heights) / len(test_heights):.2f}")
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(train_widths, bins=20, alpha=0.7, label="Widths")
plt.hist(train_heights, bins=20, alpha=0.7, label="Heights")
plt.title("Train Image Size Distribution")
plt.xlabel("Pixels")
plt.ylabel("Frequency")
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(test_widths, bins=20, alpha=0.7, label="Widths")
plt.hist(test_heights, bins=20, alpha=0.7, label="Heights")
plt.title("Test Image Size Distribution")
plt.xlabel("Pixels")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```
